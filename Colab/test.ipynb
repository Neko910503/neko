```python
import requests

def crawl_stock(stock_id):
  url = "http://www.twse.com.tw/exchangeReport/STOCK_DAY?response=json&date=20221214&stockNo=" + stock_id
  resp = requests.get(url)
  if resp.status_code != 200:
    return None
  else:
    return resp.json()

# 範例：抓取台積電的股票資料
stock_data = crawl_stock("2330")
print(stock_data)
```
